#### Data Preperation:

```python
# altering 'Object' type faetures to thier actual Dtype:
# String columns
str_features = df[['Card Type','Gender','Geography']]

for col in str_features:
    df[col] = df[col].astype('string')
    
# boolean columns
bool_cols = df[['HasCrCard','IsActiveMember','Exited','Complain']]

for col in bool_cols:
    df[col] = df[col].astype('bool')

# dropping irrelevant columns
df = df.drop(columns = ['RowNumber','CustomerId','Surname'])

df.info()
```
